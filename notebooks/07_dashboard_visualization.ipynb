## 1. Load Data ##

```python
import ast
import pandas as pd

df = pd.read_csv(
    "../data/processed/cleaned_jobs.csv"
)

df["skills"] = (
    df["skills"]
    .apply(ast.literal_eval)
)

from pathlib import Path

Path("../outputs/figures").mkdir(
    parents=True,
    exist_ok=True
)
```

## 2. Plotly Imports ##

```python
import plotly.express as px
```

## 3. Salary Distribution ##

```python
fig = px.histogram(
    df,
    x="salary_numeric",
    nbins=30
)

fig.show()

fig.write_html(
    "../outputs/figures/salary_distribution_interactive.html"
)
```

## 4. Salary by Seniority ##

```python
fig = px.box(
    df,
    x="seniority_level",
    y="salary_numeric"
)

fig.show()

fig.write_html(
    "../outputs/figures/seniority_level_distribution.html"
)
```

## 5. Top Hiring Locations ##

```python
location_df = (
    df["location"]
    .value_counts()
    .reset_index()
)

location_df.columns = [
    "location",
    "count"
]

fig = px.bar(
    location_df.head(15),
    x="location",
    y="count"
)

fig.show()

fig.write_html(
    "../outputs/figures/hiring_locations_interactive.html"
)
```

## 6. Top Skills ##

```python
skill_df = pd.read_csv(
    "../data/processed/skill_counts.csv"
)

fig = px.bar(
    skill_df.head(20),
    x="skill",
    y="count"
)

fig.show()

fig.write_html(
    "../outputs/figures/top_skills_interactive.html"
)
```
