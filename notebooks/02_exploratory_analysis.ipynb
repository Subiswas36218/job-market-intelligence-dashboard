## 1. Load Data ##

```python
import pandas as pd

df = pd.read_csv(
    "../data/processed/cleaned_jobs.csv"
)

from pathlib import Path

Path("../outputs/figures").mkdir(
    parents=True,
    exist_ok=True
)
```

## 2. Top Job Titles ##

```python
df["job_title"] \
    .value_counts() \
    .head(20)
```

## 3. Visualization ##

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

(
    df["job_title"]
    .value_counts()
    .head(10)
    .plot(kind="bar")
)

plt.title("Top 10 Job Titles")
plt.tight_layout()

plt.savefig(
    "../outputs/figures/top_job_titles.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

## 4. Top Locations ##

```python
df["location"] \
    .value_counts() \
    .head(15)
```

## 5. Salary Distribution ##

```python
plt.figure(figsize=(10,6))

df["salary_numeric"].hist(
    bins=30
)

plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Frequency")

plt.tight_layout()

plt.savefig(
    "../outputs/figures/salary_distribution.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

## 6. Seniority Distribution ##

```python
plt.figure(figsize=(10,6))

(
    df["seniority_level"]
    .value_counts()
    .plot(kind="bar")
)

plt.title("Seniority Levels")

plt.tight_layout()

plt.savefig(
    "../outputs/figures/seniority_distribution.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```
