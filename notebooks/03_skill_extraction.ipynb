## 1. Load Data ##

```python
import pandas as pd
import ast

df = pd.read_csv(
    "../data/processed/cleaned_jobs.csv"
)

from pathlib import Path

Path("../outputs/figures").mkdir(
    parents=True,
    exist_ok=True
)
```

## 2. Convert Skills String ##

```python
df["skills"] = (
    df["skills"]
    .apply(ast.literal_eval)
)
```

## 3. Count Skills ##

```python
from collections import Counter

counter = Counter()

for row in df["skills"]:
    counter.update(row)
```
