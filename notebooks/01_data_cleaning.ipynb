## 1. Import Data ##

```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/raw/data_science_job_posts_2025.csv")

print(df.shape)
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

## 2. Remove Duplicates ##

```python
df = df.drop_duplicates()

print(df.shape)
```

## 3. Fill Missing Values ##

```python
df["job_title"] = df["job_title"].fillna("Unknown")

df["seniority_level"] = df["seniority_level"].fillna("Unknown")

df["status"] = df["status"].fillna("Unknown")

df["ownership"] = df["ownership"].fillna("Unknown")

df["revenue"] = df["revenue"].fillna("Unknown")

df["location"] = df["location"].fillna("Unknown")
```

## 4. Convert Dates ##

```python
df["post_date"] = pd.to_datetime(
    df["post_date"],
    errors="coerce"
)
```

## 5. Convert Salary to Numeric Average ##

```python
import re

def salary_to_number(value):

    value = str(value)

    nums = re.findall(
        r"\d[\d,]*",
        value
    )

    nums = [
        int(x.replace(",", ""))
        for x in nums
    ]

    if len(nums) == 0:
        return np.nan

    if len(nums) == 1:
        return nums[0]

    return sum(nums)/len(nums)

df["salary_numeric"] = (
    df["salary"]
    .apply(salary_to_number)
)
```

## 7. Save Results ##

```python
df.to_csv(
    "../data/processed/cleaned_jobs.csv",
    index=False
)
```
